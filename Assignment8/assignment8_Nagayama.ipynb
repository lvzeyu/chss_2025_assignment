## 課題1

以下の図を再現してください

- 年齢の分布を、異なる階級幅を使って、4つのサブプロットをわけて描画する
   - 階級幅を適当で決めてよい
- 各サブプロットには、
    - x軸ラベル、y軸ラベル、サブプロット番号(a-d)を指定する
    - 水平方向のグリッド線を追加する
    - ヒストグラムの色を`lightblue`に変更する

![](./Figure1.png)

```python
# 課題1
fig = plt.figure(figsize=(18, 6))
ax1 =fig.add_subplot(2, 2, 1)
ax2 =fig.add_subplot(2, 2, 2)
ax3 =fig.add_subplot(2, 2, 3)
ax4 =fig.add_subplot(2, 2, 4)
ax1.hist(df["age"], bins = 100, color = "lightblue")
ax1.set_title("a")
ax1.set_xlabel("年齢（歳）")
ax1.set_ylabel("人数")

ax2.hist(df["age"], bins = 50, color = "lightblue")
ax2.set_title("b")
ax2.set_xlabel("年齢（歳）")
ax2.set_ylabel("人数")

ax3.hist(df["age"], bins = 20, color = "lightblue")
ax3.set_title("c")
ax3.set_xlabel("年齢（歳）")
ax3.set_ylabel("人数")

ax4.hist(df["age"], bins = 15, color = "lightblue")
ax4.set_title("d")
ax4.set_xlabel("年齢（歳）")
ax4.set_ylabel("人数")
```

## 課題2

以下の図を再現してください

- 男性乗客と女性乗客の年齢の分布をそれぞれ描画する
- 全乗客の年齢分布を示す灰色ヒストグラムも描画する
- 図のタイトル、x軸ラベル、y軸ラベル、凡例を設定する

![](./Figure2.png)

```python
# 課題2
fig2 = plt.figure(figsize=(18, 6))
ax_1 =fig2.add_subplot(1, 2, 1)
ax_2 =fig2.add_subplot(1, 2, 2)

male_age = df[df["sex"] == "male"]["age"]
ax_1.hist(df["age"], bins = 50, color = "gray")
ax_1.hist(male_age, bins = 50, color = "blue")
ax_1.set_title("男性乗客")
ax_1.set_xlabel("乗客の年齢（歳）")
ax_1.set_ylabel("人数")
ax_1.legend()

female_age = df[df["sex"] == "female"]["age"]
ax_2.hist(df["age"], bins = 50, color = "gray")
ax_2.hist(female_age, bins = 50, color = "red")
ax_2.set_title("女性乗客")
ax_2.set_xlabel("乗客の年齢（歳）")
ax_2.set_ylabel("人数")
ax_2.legend()

plt.show()
```
